```python
import pandas as pd
frame = pd.read_csv('C:/Users/vitor/Desktop/Estudos/DataSet.csv') 
frame
```

```python
len(frame)
```

```python
frame['Sexo'] = frame['Sexo'].replace('M','1')
frame['Sexo'] = frame['Sexo'].replace('F','0')

frame['Estado Civil'] = frame['Estado Civil'].replace("Solteiro(a)", "1")
frame['Estado Civil'] = frame['Estado Civil'].replace("Casado(a)", "0")

for index, row in frame.iterrows():
   if row['Filhos'] > 0 :
       frame.loc[index, 'Filho'] = '1'
   else:
       frame.loc[index, 'Filho'] = '0'
```

```python
del frame['Filhos']
```

```python
frame['Sexo'] = frame['Sexo'].astype(float)
frame['Estado Civil'] = frame['Estado Civil'].astype(float)
frame['Filho'] = frame['Filho'].astype(float)
```

```python
x = frame[['Sexo','Estado Civil']]
y = frame[['Filho']]
```

```python
treino_x = x[:23]
treino_y = y[:23]

teste_x = x[23:]
teste_y = y[23:]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

seed = 20

treino_x, teste_x, treino_y, teste_y = train_test_split(x, y,
                                                        random_state = 20,
                                                        test_size = 0.25,
                                                        stratify = y)

#Gera numeros aleatorios na previsao entao precisa do seed para um resultado mais preciso.
#Os numeros deixam de ser muito aleatorios com o parametro seed.
#stratify gera um proporção equivalente para o teste

modelo = LinearSVC()
modelo.fit(treino_x, treino_y.values.ravel())
previsoes = modelo.predict(teste_x)

acuracia = accuracy_score(teste_y, previsoes) * 100
print("A acurácia foi %.2f%%"  % acuracia)
```
